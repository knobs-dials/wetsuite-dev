```python
from sklearn.datasets import fetch_20newsgroups  

cat_text = {} # category_name -> [document, ...]

for category in ('alt.atheism', 'talk.religion.misc', 'sci.space','comp.sys.ibm.pc.hardware', 'comp.sys.mac.hardware', 'rec.sport.baseball', 'rec.sport.hockey'):
    bunch = fetch_20newsgroups(subset='train', categories=[category])
    cat_text[category] = bunch.data
 
```

```python
import collections

def simple_tokenize(text: str):
    ' split string into words - in a very dumb way but slightly better than just split() '
    import re
    l = re.split(r'[\s!@#$%^&*":;/~_=(){}\[\]<>,.?\xab\xbb\u2018\u2019\u201a\u201b\u201c\u201d\u201e\u201f\u2039\u203a\u2358\u275b\u275c\u275d\u275e\u275f\u2760\u276e\u276f\u2e42\u301d\u301e\u301f\uff02\U0001f676\U0001f677\U0001f678-]+', text)
    return list(e.strip("'")   for e in l  if len(e)>0)            


analysis_per_category = {} # category -> unique words used in it

for category, documents in cat_text.items():
    text = ('\n\n'.join(documents))
    #tfidf.add_document( text, category )


    terms=[]
    for doc in documents:
        header, body = doc.split('\n\n',1) # take off newsgroupy header
        wc = collections.defaultdict(int)
        for token in simple_tokenize(body):
            wc[token.lower()] += 1
        for term, count in wc.items():
            if count >= 3:
                terms.append( term )

    analysis_per_category[category] = terms
    
```

```python
def unique_in_lists( dd ):
    """ given a  {cat -> list-of-term}  dict,
        returns  {cat -> terms-not-in-any-other-catlist} 
    """
    ret = {}
    keys = list(dd.keys())
    for okey in keys: # feels like there's a clearer way of doing this, but it works
        this, others = set(), set()
        for ikey in keys:
            if okey==ikey: # 
                this.update( dd[ikey] )
            else:
                others.update( dd[ikey] )
        ret[ okey ] = list(this.difference(others))
    return ret
    # version that takes a list of lists:
    # ret = []
    # lllen = len(ll)
    # for o in range(lllen):
    #     this, others = set(), set()
    #     for i in range(lllen):
    #         if o==i: # 
    #             this.update( ll[i] )
    #         else:
    #             others.update( ll[i] )
    #     ret.append( list(this.difference(others)) ) 
    # return ret


#allterms = {}
#or category, terms in analysis_per_category.items():
#    tops[category] = top

#
for category, uu in unique_in_lists( analysis_per_category ).items():
    print( '%25s  %s'%( category, uu) )
```
